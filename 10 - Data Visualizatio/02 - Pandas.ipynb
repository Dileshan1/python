```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Numerical data

# Line chart

```python
df=pd.DataFrame({"Year":[1990,1991,1992,1993,1994,1995,1996,1997,1998,1999,2000],
                 "Weight":[76,64,65,70,78,79,75,74,69,68,65],
                "Height":[167,168,168,168,169,170,171,171,172,173,174]})
df
```

```python
df.plot.line(y="Weight",figsize=(8,4))
plt.xlabel("Index")
plt.ylabel("Weight")
plt.title("Weight with years")
plt.show()
```

```python
df["Weight"].plot.line(figsize=(8,4))           # as a time series

plt.show()
```

```python
df=pd.DataFrame({"Col01":np.random.random(100),"Col02":np.random.randn(100)})
df.head()

df.plot.line(y="Col01")
plt.show()
```

```python
df["Col01"].plot.line()         # same
plt.show()
```

# Histogram

```python
df=pd.DataFrame({"Col01":np.random.random(100),"Col02":np.random.randn(100)})
df.head()
```

```python
df["Col02"].hist()
plt.xlabel("Col02")
plt.ylabel("Frequency")
plt.title("Col02 Histogram")
plt.show()
```

```python
df.plot.hist(y="Col01")
plt.xlabel("Col01")
plt.ylabel("Frequency")
plt.title("Col01 Histogram")
plt.show()
```

# Boxplot 

```python
df=pd.DataFrame({"Col01":np.random.random(100),"Col02":np.random.random(100)})
df.head()
```

```python
df["Col01"].plot.box()
plt.title("Col01 Boxplot")
plt.show()
```

```python
df.plot.box()
plt.title("Boxplot for all columns")
plt.show()
```

```python
df.boxplot()          # same
plt.show()
```

# Categorical data

# Bar graphs

```python
data = {"City":["Kandy", "Colombo", "Galle"], "Visits":[50,45,70]}
df=pd.DataFrame(data)
df
```

```python
df.plot.bar(x="City", y="Visits")
plt.title("Tourist visits during last week")
plt.ylabel("Frequency")
plt.show()
```

```python
df.plot.barh(x="City", y="Visits")
plt.title("Tourist visits during last week")
plt.ylabel("Frequency")
plt.show()
```

# Pie charts

```python
data = {"Visits":[50,45,70]}
df=pd.DataFrame(data,index=["Kandy", "Colombo", "Galle"])
df
```

```python
df.plot.pie(y="Visits",figsize=(8,10))
plt.title("Tourist visits during last week")
plt.ylabel("Frequency")
plt.show()
```

# Categorical VS Categorical data

# Multiple bar graphs

```python
df=pd.DataFrame({"Maths":[90,60],"Science":[78,83]},index=["Male","Female"])
df
```

```python
df.plot.bar()
plt.title("Maths & Science marks with gender")
plt.ylabel("Frequency")
plt.show()
```

```python
df.plot.barh()
plt.title("Maths & Science marks with gender")
plt.ylabel("Frequency")
plt.show()
```

# Stacked bar graphs

```python
df.plot.bar(stacked=True)
plt.title("Maths & Science marks with gender")
plt.ylabel("Frequency")
plt.show()
```

```python
df.plot.barh(stacked=True)
plt.title("Maths & Science marks with gender")
plt.ylabel("Frequency")
plt.show()
```

# 100% Stacked bar graphs

```python
df=pd.DataFrame({"Maths":[90,60],"Science":[78,83]},index=["Male","Female"])
df
```

```python
df_perc=df.div(df.sum(axis=1),axis=0)*100
df_perc
```

```python
df_perc.plot.bar(stacked=True)
plt.title("Maths & Science marks with gender")
plt.ylabel("Frequency")
plt.show()
```

```python
df_perc.plot.barh(stacked=True)
plt.title("Maths & Science marks with gender")
plt.ylabel("Frequency")
plt.show()
```
