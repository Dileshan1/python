# Importing matplotlib.pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# A simple line chart

```python
L1=[1,2,3,4]
L2=[10,25,32,58]
plt.plot(L1,L2)
plt.show()
```

# Adding a title and axis names

```python
L1=[1,2,3,4]
L2=[10,25,32,58]
plt.plot(L1,L2)
plt.title("A Simple Line Plot")
plt.xlabel("L1")
plt.ylabel("L2")
plt.show()
```

```python
L1=[1,2,3,4]
L2=[10,25,32,58]
plt.figure(figsize=(8,7))
plt.plot(L1,L2)
plt.title("A Simple Line Plot")
plt.xlabel("L1")
plt.ylabel("L2")
plt.show()
```

# Subplots in the same figure

# First method

```python
x=np.arange(10)
y1=x+5
y2=x**2
plt.figure(figsize=(10,5))

#Plots are created with a single row and two columns
plt.subplot(1,2,1)
plt.plot(x,y1,"b*")
plt.title("First chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(1,2,2)
plt.plot(x,y2,"go")
plt.title("Second chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
x=np.arange(10)
y1=x+5
y2=x**2
y3=-x
y4=np.sqrt(x)
plt.figure(figsize=(15,10))

#Plots are created with a two rows and two columns
plt.subplot(2,2,1)
plt.plot(x,y1,"b*")
plt.title("First chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(2,2,2)
plt.plot(x,y2,"go")
plt.title("Second chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(2,2,3)
plt.plot(x,y3,"r+")
plt.title("Third chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(2,2,4)
plt.plot(x,y4,"k-")
plt.title("Fourth chart")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

# Second method

```python
x=np.arange(10)
y1=-(x**2)
y2=x**3
y3=np.exp(x)
y4=np.sqrt(x)

#Plots are created with a two rows and two columns
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,10))

ax[0,0].plot(x,y1,"b*")
ax[0,0].set_title("First chart")
ax[0,0].set_xlabel("X")
ax[0,0].set_ylabel("Y")

ax[0,1].plot(x,y2,"go")
ax[0,1].set_title("Second chart")
ax[0,1].set_xlabel("X")
ax[0,1].set_ylabel("Y")

ax[1,0].plot(x,y3,"r+")
ax[1,0].set_title("Thrird chart")
ax[1,0].set_xlabel("X")
ax[1,0].set_ylabel("Y")

ax[1,1].plot(x,y4,"k-")
ax[1,1].set_title("Fourth chart")
ax[1,1].set_xlabel("X")
ax[1,1].set_ylabel("Y")


plt.show()
```

# Pie charts with percentages

```python
Gender=["Male","Female"]
Number_Of_Students=[45,75]

plt.pie(Number_Of_Students,labels=Gender,autopct="%.4f",shadow=True)
plt.legend(title="Gender")
plt.axis("equal")
plt.title("Pie of gender")
plt.show()
```

```python
Gender=["Male","Female"]
Number_Of_Students=[45,75]

plt.pie(Number_Of_Students,labels=Gender,autopct="%1.0f%%")        # This gives the percentage symbol
plt.legend(title="Gender")
plt.axis("equal")
plt.title("Pie of gender")
plt.show()
```

# Scatter plot with displaying the categories in another variable

```python
cls = np.random.choice(a=["a", "c", "b"],size=100,p=[0.5, 0.3, 0.2])
height=np.random.normal(160,5,100)
weight=np.random.normal(55,2,100)
col=["red" if i=="a" else "green" if i=="b" else "blue" for i in cls]


plt.scatter(weight,height,c=col)
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("Height VS Weight")
plt.show()
```

# Side by side boxplots

```python
cls = np.random.choice(a=["a", "c", "b"],size=100)
height=np.random.normal(160,5,100)

df=pd.DataFrame({"Height":height,"Class":cls})
df.head()
```

```python
L_new=[list(df[df["Class"]==i]["Height"]) for i in ["a", "c", "b"]]
L_new
```

```python
plt.boxplot(L_new,positions=[0,2,4])
plt.xticks([0,2,4],["a","b","c"])
plt.show()
```

# 3D Scatter plot

```python
from mpl_toolkits import mplot3d
```

```python
height=np.random.normal(160,5,100)
weight=np.random.normal(55,2,100)
age=np.random.randint(24,45,100)

plt.figure(figsize=(10,10))
ax=plt.axes(projection="3d")
ax.scatter3D(weight,height,age)

ax.set_xlabel("Weight")
ax.set_ylabel("Height")
ax.set_zlabel("Age")

plt.title("Relationship among Weight, Height & Age")
plt.show()
```
