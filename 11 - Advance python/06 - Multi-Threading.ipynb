```python
class A():
    def process(self):
        for i in range(5):
            print("AAA")
    
class B():
    def process(self):
        for i in range(5):
            print("BBB")
```

```python
ob1=A()
ob2=B()
```

```python
ob1.process()
ob2.process()
```

```python
from threading import Thread

class A(Thread):
    def run(self):
        for i in range(100):
            print("AAA")
    
class B(Thread):
    def run(self):
        for i in range(100):
            print("BBB")
```

```python
ob1=A()
ob2=B()
```

```python
ob1.start()
ob2.start()
```

```python
from threading import Thread
from time import sleep

class A(Thread):
    def run(self):
        for i in range(10):
            print("AAA")
            sleep(1)
    
class B(Thread):
    def run(self):
        for i in range(10):
            print("BBB")
            sleep(1)
```

```python
ob1=A()
ob2=B()
```

```python
ob1.start()
ob2.start()
```

```python
ob1=A()
ob2=B()

ob1.start()
sleep(0.2)         # This is for removing the collision
ob2.start()
```

```python
ob1=A()
ob2=B()

ob1.start()
sleep(0.2)
ob2.start()

print("CCC")
```

```python
ob1=A()
ob2=B()

ob1.start()
sleep(0.2)
ob2.start()

ob1.join()
ob2.join()

print("CCC")
```
